```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of 
#        each plot.
```

```python
# Ans- Five plots :-
        # Scatter Plot - A scatter plot is used to visualize the relationship between two variables. Seaborn's 
        #                scatterplot() function provides a simple way to create scatter plots.
        
        # Line Plot -  A line plot is used to visualize the trend of a variable over time. Seaborn's lineplot() 
        #              function provides a simple way to create line plots.
        
        # Histogram -  A histogram is used to visualize the distribution of a variable. Seaborn's histplot() 
        #              function provides a simple way to create histograms.
        
        # Violin Plot - A violin plot is similar to a box plot, but provides a more detailed view of the 
        #               distribution of the data. Seaborn's violinplot() function provides a simple way to create
        #               violin plots.
        
        # Pairplot - A pairplot is used to visualize the relationship between multiple variables. Seaborn's 
        #            pairplot() function provides a simple way to create pairplots.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot 
#        using x = "timepoint" and y = "signal" for different events and regions.
#  Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x = 'timepoint', y = "signal", data = fmri)
```

```python
fmri['event'].value_counts()
```

```python
sns.lineplot(x = 'timepoint', y = "signal", data = fmri, hue = 'event')
```

```python
fmri['region'].value_counts()
```

```python
sns.lineplot(x = 'timepoint', y = "signal", data = fmri, hue='region')
```

```python
sns.lineplot(x = 'timepoint', y = "signal", data = fmri, hue='region', style='event')
```
